```python
def discount(price, customer_type):
    discount = 0
    if customer_type == "Regular":
        discount = 5
    if customer_type == "Premium":
        discount = 15
    if customer_type == "Employee":
        discount = 25

    final_price = price - (price * discount / 100)
    return final_price

print(" Product Discount Evaluation ")
name = input("Enter Customer Name: ")
price = float(input("Enter Product Price: "))
customer_type = input("Enter Customer Type (Regular/Premium/Employee): ")

final = discount(price, customer_type)

print("Customer Name:", name)
print("Original Price: Rs.", price)
print("Customer Type:", customer_type)
print("Final Price After Discount: Rs.", final)
```
