```python
def password_strength(password):
    if len(password) < 8:
        print("Reason: Password is less than 8 characters.")
        return "Weak"
    i = 0
    digit = 0
    special = 0
    while i < len(password):
        ch = password[i]
        if ch >= "0" and ch <= "9":
            digit = 1
        if ch == "!" or ch == "#" or ch == "@" or ch == "$":
            special = 1
        i = i + 1

    if digit == 0:
        print("Reason: No digit found in password.")
        return "Weak"
    if special == 0:
        print("Reason: No special character found in password.")
        return "Weak"

    print("Reason: Password meets all requirements.")
    return "Strong"

print(" Password Strength Evaluation ")
name = input("Enter User Name: ")
print(" Password included Alpabets, Numbers, Special Character ")
password = input("Enter Password: ")

status = password_strength(password)

print("User Name:", name)
print("Password Strength Status:", status)
```
