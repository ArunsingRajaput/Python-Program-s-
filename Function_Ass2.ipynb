```python
def check_attendance(login_time):
    if login_time == "9:30":
        return "Present"
    elif login_time <= "10:00":
        return "Late"
    else :
        return "Absent"

print(" Attendance Status Checker ")
name = input("Enter Name : ")
login_time = input("Enter Login Time (HH:MM) : ")
status = check_attendance(login_time)
print("Name : ", name)
print("Status : ", status)
```
