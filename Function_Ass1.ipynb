```python
def eligibility(age, percentage):
    if age >= 18 and percentage >= 60:
        print("\nReason : Age must be 18 or above and percentage must be 60% or above.")
        return "Eligible"
    else:
        if age < 18:
            print("\nReason : Age is below 18.")
        if percentage < 60:
            print("\nReason : Graduation percentage is below 60%.")
        return "Not Eligible"

print(" Intern Eligibility Evaluation Function. ")
name = input("Enter Name : ")
age = int(input("Enter Age : "))
percentage = float(input("Enter Percentage : "))
status = eligibility(age, percentage)
print("Name : ", name)
print("Age : ", age)
print("Percentage : ", percentage)
print("Status : ", status)
```
