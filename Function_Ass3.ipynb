```python
def ele_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 2
    if units > 100 and units <= 200:
        bill = units * 4
    if units > 200:
        bill = units * 6
    return bill

print(" Electricity Bill Calculation ")
name = input("Enter Customer Name: ")
units = int(input("Enter Units : "))
bill = ele_bill(units)
print("Customer Name : ", name)
print("Units : ", units)
print("Total Bill Amount : Rs.", bill)
```
