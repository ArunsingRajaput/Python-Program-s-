```python
print("Course Fee Billing System.")
python_fee = 5000
data_fee = 8000
ai_ml_fee = 12000

course = input("Enter Course Name (Python Programming / Data Analytics / AI & ML): ")

if course == "Python Programming":
    fee = python_fee
elif course == "Data Analytics":
    fee = data_fee
elif course == "AI & ML":
    fee = ai_ml_fee
else:
    print("Invalid Course Selected")
    exit()

student = input("Are you a student? (yes/no): ")
early = input("Is this early registration? (yes/no): ")

student_discount = 0.10 * fee if student == "yes" else 0
early_discount = 0.05 * fee if early == "yes" else 0

total_discount = student_discount + early_discount
final_amount = fee - total_discount

print("\n-- COURSE FEE BILL --")
print("Course Name         :", course)
print("Original Fee (₹)    :", fee)
print("Student Discount    :", student_discount)
print("Early Reg Discount  :", early_discount)
print("-------------------------------------")
print("Total Discount (₹)  :", total_discount)
print("Final Payable (₹)   :", final_amount)
print("------------------------------------")
```
